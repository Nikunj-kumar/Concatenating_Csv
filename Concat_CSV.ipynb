```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing Csv into pandas dataframe
df1 = pd.DataFrame(pd.read_csv("Desktop/777_m4_datasets_v1.0/PhysicsScoreTerm1.csv"))
df2 = pd.DataFrame(pd.read_csv("Desktop/777_m4_datasets_v1.0/MathScoreTerm1.csv"))
df3 = pd.DataFrame(pd.read_csv("Desktop/777_m4_datasets_v1.0/DSScoreTerm1.csv"))
```

```python
# concatenating all file into one named as ScoreFinal
df4 = pd.concat([df1,df2,df3])
df4.to_csv("Desktop/777_m4_datasets_v1.0/ScoreFinal.csv")
df4.head()
```

```python
# Replacing blanks Score with 0.
df4['Score'].fillna(0, inplace=True)
```

```python
# Dropping name and ethnicity to avoid discrimination on these basis.
df4.drop(['Name','Ethinicity'], axis=1, inplace=True)
df4.head()
```

```python
# Assigning male(M) as 1 and female(F) as 2 
rep  = {'M':1,'F':2}
df4['Sex'] = df4['Sex'].map(rep)
df4.head()
```

```python
#Saving all update thing into the file
df4.to_csv("Desktop/777_m4_datasets_v1.0/ScoreFinal.csv")
df4
```
